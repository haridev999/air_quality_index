```python
dfnull=pd.DataFrame(dff[['PM10','NH3','Benzene','Toluene','Xylene']])
dfnull
```

```python
data=dff.drop(columns=['PM10','NH3','Benzene','Toluene','Xylene'])
data.isna().sum()
```

```python
data=data.dropna()
data.isna().sum()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.isna(),cmap='viridis')
plt.show()
```

```python
data.duplicated().sum()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data)
plt.yscale('log')
```

```python
for col in data.select_dtypes(include=['float64','int64']).columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    data[col+'_outlier'] = ((data[col] < lower) | (data[col] > upper))
data.head()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(data)
plt.xticks(rotation=90)
```
